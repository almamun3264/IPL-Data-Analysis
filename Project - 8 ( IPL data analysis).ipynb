# Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### loading csv file

```python
data = pd.read_csv('matches.csv')
```

```python
data.head()
```

# 3. Getting the freequency of most man of the match award.

```python
data['player_of_match'].value_counts()
```

# 4. Getting top 10 frequency of man of the match and visualize it.

```python
data['player_of_match'].value_counts()[0:10]
```

### visualize
### *bar plot*

```python
plt.figure(figsize = (16,6))
plt.title('Man of the match award' , color = 'r', size = 20)
plt.xlabel('Name',color = 'g',size = 16)
plt.ylabel('Award ',color = 'g',size = 16)
plt.bar(data['player_of_match'].value_counts()[0:10].keys() , data['player_of_match'].value_counts()[0:10],color = ["#FF0000","#00FF00","#0000FF","#FFFF00","#00FFFF","#FF4500","#F08080","#B8860B","#6B8E23","#00FA9A"])
plt.show()
```

# 5. Getting the frequency of result column.

```python
data['result'].value_counts()
```

# 6. Find the number of toss win every team .

```python
data['toss_winner'].value_counts()
```

### visualize

```python
plt.figure(figsize = (16,6))
plt.title('Toss winning chances of each team' , color = 'r', size = 20)
plt.xlabel(' Team Name',color = 'g',size = 16)
plt.ylabel('Toss win ',color = 'g',size = 16)
plt.xticks(rotation = 75)
plt.bar(data['toss_winner'].value_counts().keys() , data['toss_winner'].value_counts())
plt.show()
```

# 7. Chance of wining of a team batting first.

```python
batting_first = data[data['win_by_runs'] != 0]
```

```python
batting_first.head()
```

```python
batting_first['winner'].value_counts()
```

### visualize
### *bar plot*

```python
plt.figure(figsize = (16,6))
plt.title('Win by runs | Batting first' , color = 'r', size = 28)
plt.xlabel(' Team Name',color = 'g',size = 20)
plt.ylabel('Win match ',color = 'g',size = 20)
plt.xticks(rotation = 65)
plt.bar(batting_first['winner'].value_counts().keys() ,batting_first['winner'].value_counts())
plt.show()
```

### *pie chart*

```python
plt.figure(figsize = (7,7))
plt.title('Win by runs | Batting first' , color = 'r', size = 28)
plt.pie(list(batting_first['winner'].value_counts()) , labels = list(batting_first['winner'].value_counts().keys()) ,autopct = '%0.1f%%')
plt.show()
```
